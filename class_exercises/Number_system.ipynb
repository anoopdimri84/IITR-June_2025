<span style="background-color:yellow;">**Q.6 Convert base10 to base5**.</span>
- For base10, first digit + 10 * second digit + 100 * third digit....
- For base5, first digit +5 * second digit + 5^2 * third digit....

```python
# Q.6 Basic approach - Step-by-step approach
num = int(input("Put your two digit number in base10: "))
first_digit = num % 5 # as we are converting to base5
print("first digit: ", first_digit)
remaining_digit = num-first_digit
num=remaining_digit//5
second_digit = num % 5
print("second digit: ", second_digit)
remaining_digit = num-second_digit
num=remaining_digit//5
print("third digit: ", num)
print(f"number in base10: {num}{second_digit}{first_digit}")
```

```python
# Q.6 Programming approach
num = int(input("Put your number in base10: "))
base5_num = []
inputnum = num
while inputnum >= 5:
    digit = inputnum % 5
    base5_num.append(digit)
    inputnum = inputnum // 5
base5_num.append(inputnum)
base5_num.reverse()
print(base5_num)
```

<span style="background-color:yellow;">**Q.7 Convert base N to base M**.</span>

```python
num = int(input("Put your number in base10: "))
base = int(input("Put your base: "))
base_num = []
inputnum = num
while inputnum >= base:
    digit = inputnum % base
    base_num.append(digit)
    inputnum = inputnum // base
base_num.append(inputnum)
base_num.reverse()
print(base_num)
```
